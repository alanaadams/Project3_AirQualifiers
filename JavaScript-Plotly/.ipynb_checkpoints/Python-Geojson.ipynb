```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import csv, json
from pprint import pprint
import urllib
from geojson import Feature, FeatureCollection, Point
```

```python
features = []
with open('measurements_north_america.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)
    for parameter,value,lastUpdated,unit,location,country,coordinates,latitude,longitude in reader:
        latitude, longitude = map(float, (latitude, longitude))
        features.append(
              Feature(
                geometry = Point(( latitude, longitude)),
                properties = {
                    'parameter': parameter,
                    'value': value,
                    'lastUpdated' : lastUpdated,
                     'unit': unit,
                     'location':location,
                     'country':country,
                     'coordinates':coordinates
                }
            )
        )
collection = FeatureCollection(features)
with open("measurementNA.json", "w", encoding='utf-8') as f:
    f.write('%s' % collection)
```

```python
pprint(collection);
```
