```python
import pandas as pd
df = pd.DataFrame(data['results'])

df['latitude'] = df["coordinates"].apply(lambda x: x['latitude'])
df['longitude'] = df["coordinates"].apply(lambda x: x['longitude'])
# df.drop('coordinates', axis=1, inplace=True)

# https://stackoverflow.com/questions/49671693/pandas-dataframe-normalize-one-json-column-and-merge-with-other-columns
df1 = (pd.concat({i: pd.DataFrame(x) for i, x in df.pop('measurements').items()})
         .reset_index(level=1, drop=True)
         .join(df)
         .reset_index(drop=True))

df1 = df1.drop('city',axis=1)

df1.to_csv('measurements_flattened.csv', index=False)
```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import scipy.stats as st

file_path = "C:/Users/cmfjr/Desktop/class/Week15/measurements_north_america.csv"
measure_df = pd.read_csv(file_path)

pm25_df = measure_df[measure_df['parameter'] == 'pm25']
humidity_df = measure_df[measure_df['parameter'] == 'humidity']


merged_df = pd.merge(humidity_df,pm25_df, on='coordinates')
merged_df = merged_df[merged_df['value_y'] < 50]

pm25_list = merged_df['value_y'].tolist()
humidity_list = merged_df['value_x'].tolist()

plt.scatter(pm25_list, humidity_list);
plt.xlabel('PM25');
plt.ylabel('Humidty (%)');
plt.title('PM25 v Humidty (%)')
#Build line of best fit
slope, intercept, rvalue, pvalue, stderr = st.linregress(pm25_list, humidity_list)
x = merged_df['value_y']
y = slope * x + intercept
r_value_1 = "R = " + str(round(rvalue,2))
plt.plot(x, y, color="red")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(30,95),fontsize=15,color="black")
plt.annotate(r_value_1,(30,85),fontsize=15,color="black")

plt.savefig('scatter_plot1.png')
```

```python
pm25_df = measure_df[measure_df['parameter'] == 'pm25']
temperature_df = measure_df[measure_df['parameter'] == 'temperature']

merged_df = pd.merge(temperature_df,pm25_df, on='coordinates')
merged_df = merged_df[merged_df['value_y'] < 50]
merged_df = merged_df[merged_df['value_x'] > 5]

pm25_list = merged_df['value_y'].tolist()
temperature_list = merged_df['value_x'].tolist()

plt.scatter(pm25_list, temperature_list);
plt.xlabel('PM25');
plt.ylabel('Temperature (F)');
plt.title('PM25 v Temperature (F)')

#Build line of best fit
slope, intercept, rvalue, pvalue, stderr = st.linregress(pm25_list, temperature_list)
x = merged_df['value_y']
y = slope * x + intercept
r_value_1 = "R = " + str(round(rvalue,2))
plt.plot(x, y, color="red")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(30,115),fontsize=15,color="black")
plt.annotate(r_value_1,(30,105),fontsize=15,color="black")

plt.savefig('scatter_plot2.png')
```
